```python
import pandas as pd

df=pd.read_csv("Car_sales.csv")
df.head()
```

```python
df.dropna()
df=df.fillna(0)
```

```python
df['Latest_Launch'] = pd.to_datetime(df['Latest_Launch'])
```

```python
df['Latest_Launch']=df['Latest_Launch'].dt.strftime('%d-%m-%Y')
```

```python
df.head()
```

```python
df=df.rename(columns={'__year_resale_value':'Resale_value'})
```

```python
df.to_csv('Cleaned_Car_sales.csv',index=False)
```
